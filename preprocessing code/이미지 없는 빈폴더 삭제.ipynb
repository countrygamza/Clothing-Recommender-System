```python
import glob
import shutil

folder_paths=glob.glob("./여성_하의_청바지/*")#삭제할 빈폴더명 입력
for folder_path in folder_paths:
    image_path = glob.glob(folder_path+"/*.png") #png로 끝나는 파일이 없으면 삭제
    if not image_path:
        print(folder_path+" image is empty")
        shutil.rmtree(folder_path)   
```
