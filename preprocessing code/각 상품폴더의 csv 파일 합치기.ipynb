```python
import glob

path = r'./test1/*'
all_files = glob.glob(path + "/*.csv")

print ("all_files: {}".format(all_files))

file_list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, encoding = 'utf-8')
    file_list.append(df)
    
pd.concat(file_list, axis=0, ignore_index=True)
```

```python
aaa= pd.DataFrame(pd.concat(file_list, axis=0, ignore_index=True))
```
